### Question 1: Import Data and Store-wise Revenue Analysis (4 Marks)

Import `pandas` and `numpy`. Download the dataset `sales.csv` from https://drive.google.com/drive/folders/1gieHICVDBbUKMZiSF4YRQMAJic-w50JM?usp=sharing. Load the dataset `sales.csv` into a DataFrame. Then, find the top 5 stores with the highest total sales and the bottom 5 stores with the lowest total sales.

**Tasks:**
- Aggregate total sales by store_id
- Rank stores by total revenue
- Compare average sales per department within those stores

```python
# Write your answer here

import pandas as pd

# Load dataset:
df = pd.read_csv("sales.csv")

# Aggregate Total sales by store:
store_sales = df.groupby('store_id')['weekly_sales'].sum()

# Ranking stores by revenue:
store_sales=store_sales.sort_values(ascending=False)

# Top 5 stores:
top_5_stores = store_sales.head(5)
print("Top 5 Stores:")
print(top_5_stores)

# Bottom 5 stores:
bottom_5_stores = store_sales.tail(5)
print("\nBottom 5 Stores:")
print(bottom_5_stores)

# Department_wise average sales for Top 5 stores: 
top_dept_avg = df[
    df['store_id'].isin(top_5_stores.index)
].groupby(
    ['store_id', 'department']
)['weekly_sales'].mean()
print("\nAverage Department Sales for Top 5 Stores:")
print(top_dept_avg)

# Department_wise average sales for Bottom 5 stores: 
bottom_dept_avg = df[
    df['store_id'].isin(bottom_5_stores.index)
].groupby(
    ['store_id', 'department']
)['weekly_sales'].mean()
print("\nAverage Department Sales for Bottom 5 Stores:")
print(bottom_dept_avg)



```

### Question 2: Department Performance Consistency (3 Marks)

Which departments have the most stable weekly sales and which are the most volatile?

**Tasks:**
- Use standard deviation and coefficient of variation
- Identify highly stable and highly volatile departments

```python
# Write your answer here

# Calculate standard deviation of weekly sales for each department:
dept_std = df.groupby('department')['weekly_sales'].std().reset_index()
dept_std.columns = ['department', 'std_dev']
print("Department-wise Standard Deviation:")
print(dept_std)

# Calculate average weekly sales for each department:
dept_mean = df.groupby('department')['weekly_sales'].mean().reset_index()
dept_mean.columns = ['department', 'mean_sales']
print("\nDepartment-wise Mean Sales:")
print(dept_mean)

# Merge standard deviation and mean sales data:
dept_cv = dept_std.merge(dept_mean, on='department')
print("\nMerged Data:")
print(dept_cv)

# Calculate Coefficient of Variation (CV):
dept_cv['cv'] = (dept_cv['std_dev'] / dept_cv['mean_sales']) * 100
print("\nDepartment-wise Coefficient of Variation:")
print(dept_cv)

# Find the most stable department (lowest CV):
most_stable = dept_cv.loc[dept_cv['cv'].idxmin()]
print("\nMost Stable Department:")
print(most_stable)

# Find the most volatile department (highest CV):
most_volatile = dept_cv.loc[dept_cv['cv'].idxmax()]
print("\nMost Volatile Department:")
print(most_volatile)
```

### Question 3: Holiday vs Non-Holiday Sales (3 Marks)

Analyze whether holidays significantly affect weekly sales.

**Tasks:**
- Compare average sales during holidays vs non-holidays
- Calculate percentage increase/decrease
- Identify departments benefiting most from holidays

```python
# Write your answer here

# Compare average sales during holidays vs non-holidays:
holiday_sales = df.groupby('is_holiday')['weekly_sales'].mean().reset_index()
holiday_sales.columns = ['is_holiday', 'avg_sales']

# Replace 0 and 1 with readable labels:
holiday_sales['is_holiday'] = holiday_sales['is_holiday'].replace({
    0: 'Non-Holiday',
    1: 'Holiday'
})
print(holiday_sales)

# Calculate percentage increase in sales during holidays:

non_holiday = holiday_sales.loc[
    holiday_sales['is_holiday'] == 'Non-Holiday',
    'avg_sales'
].iloc[0]

holiday = holiday_sales.loc[
    holiday_sales['is_holiday'] == 'Holiday',
    'avg_sales'
].iloc[0]

pct_change = ((holiday - non_holiday) / non_holiday) * 100
print(f"\nPercentage Increase = {pct_change:.2f}%")

# Calculate average holiday sales for each department:

holiday_dept = df[df['is_holiday'] == 1] \
    .groupby('department')['weekly_sales'] \
    .mean() \
    .reset_index()
holiday_dept.columns = ['department', 'holiday_avg_sales']
print("\nDepartment-wise Holiday Sales:")
print(holiday_dept)

# Identify department benefiting most from holidays:
best_dept = holiday_dept.loc[
    holiday_dept['holiday_avg_sales'].idxmax()
]
print("\nDepartment Benefiting Most From Holidays:")
print(best_dept)
```

### Question 4: Seasonal Trend Detection (3 Marks)

Identify whether sales increase or decrease during specific months.

**Tasks:**
- Extract month from the date column
- Compute average monthly sales
- Determine highest and lowest sales months

```python
# Write your answer here

# Convert date column to datetime format:
df['date'] = pd.to_datetime(df['date'])

# Extract month from date column:
df['month'] = df['date'].dt.month

# Calculate average monthly sales:
monthly_sales = df.groupby('month')['weekly_sales'].mean().reset_index()

# Rename columns:
monthly_sales.columns = ['month', 'avg_sales']

# Display average monthly sales:
print(monthly_sales)

# Find month with highest average sales:
highest_month = monthly_sales.loc[monthly_sales['avg_sales'].idxmax()]

# Find month with lowest average sales:
lowest_month = monthly_sales.loc[monthly_sales['avg_sales'].idxmin()]

# Display results:
print("\nHighest Sales Month:")
print(highest_month)

print("\nLowest Sales Month:")
print(lowest_month)
```

### Question 5: Store and Department Contribution (4 Marks)

Which combinations of store and department contribute the most to overall revenue?

**Tasks:**
- Group using store_id and department
- Find top and worst-performing combinations
- Compute percentage contribution

```python
# Write your answer here

# Group data by store_id and department and calculate total sales:
store_dept_sales = df.groupby(
    ['store_id', 'department']
)['weekly_sales'].sum().reset_index()

store_dept_sales.columns = [
    'store_id',
    'department',
    'total_sales'
]
print(store_dept_sales)

# Find top-performing and worst-performing store-department combinations:
top_combination = store_dept_sales.loc[
    store_dept_sales['total_sales'].idxmax()
]

worst_combination = store_dept_sales.loc[
    store_dept_sales['total_sales'].idxmin()
]
print("\nTop Performing Combination:")
print(top_combination)

print("\nWorst Performing Combination:")
print(worst_combination)

# Calculate total revenue from all store-department combinations:
total_revenue = store_dept_sales['total_sales'].sum()
print("\nTotal Revenue =", total_revenue)

# Compute percentage contribution of top and worst combinations:
top_contribution = (
    top_combination['total_sales'] / total_revenue
) * 100

worst_contribution = (
    worst_combination['total_sales'] / total_revenue
) * 100
print(f"\nTop Combination Contribution = {top_contribution:.2f}%")
print(f"Worst Combination Contribution = {worst_contribution:.2f}%")
```

### Question 6: Load MNIST Data (3 Marks)

Load the MNIST dataset using `sklearn.datasets.load_digits`. Separate the dataset into features (`X`) and target labels (`y`).
Print the shape of the features and the target arrays.

**Expected Output:** The shape of `X` and `y`.

```python
# Write your answer here

# Question 6: Load MNIST (Digits) Dataset:
from sklearn.datasets import load_digits
digits = load_digits()

X = digits.data
y = digits.target

print("Shape of X =", X.shape)
print("Shape of y =", y.shape)
```

### Question 7: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the MNIST features (`X`). 
Since we know there are 10 digits (0-9), set the number of clusters to 10. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
# Write your answer here

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42)
kmeans.fit(X)
kmeans_labels = kmeans.labels_

print("Shape of cluster labels =", kmeans_labels.shape)
print(kmeans_labels[:20])
```

### Question 8: F1 Score Evaluation for Clustering (5 Marks)

Evaluate the clustering performance using the F1 score. Since K-Means assigns arbitrary cluster labels, you will first need to map each cluster label to the most frequent true label in that cluster. 
After mapping the labels, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
# Write your answer here

# F1 Score Evaluation for Clustering:

from sklearn.metrics import f1_score
from scipy.stats import mode
import numpy as np

mapped_labels = np.zeros_like(kmeans_labels)
for i in range(10):
    mask = (kmeans_labels == i)
    if np.sum(mask) > 0:
        mapped_labels[mask] = mode(y[mask], keepdims=True).mode[0]
f1 = f1_score(y, mapped_labels, average='macro')

print("Macro F1 Score =", f1)
```
